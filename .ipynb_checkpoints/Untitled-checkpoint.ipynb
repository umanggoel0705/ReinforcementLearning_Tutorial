### 1. Import Dependencies

```python
import os
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

### 2. Load Environment

```python
environment_name = 'CartPole-v1'
env = gym.make(environment_name, render_mode='human')
```

```python
episodes = 5
for episode in range(episodes):
    state = env.reset()
    done = False
    truncate = False
    score = 0

    while not done and not truncate:
        # env.render()
        action = env.action_space.sample()
        n_state, reward, done, truncate, info = env.step(action)
        score += reward
    print(f"Episode:{episode+1} Score:{score}")
env.close()
```

### Understanding the Environment

```python
env.action_space
```

##### Two values:
- 0: Push the cart to left
- 1: Push the cart to right

```python
env.action_space.sample()
```

```python
env.observation_space
```

##### Four Values:
- 0: Cart Position
- 1: Cart Velocity
- 2: Pole Angle
- 3: Pole Angular Velocity

```python
env.observation_space.sample()
```

### 3. Training an RL Model

```python
# Make the directories first
log_path = "Logs"
```

```python
log_path
```

```python
env = gym.make(environment_name, render_mode='human')
env = DummyVecEnv([lambda : env])
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=5000)
env.close()
```

### 4. Save and Reload Model

```python
PPO_Path = os.path.join('Saved Models', 'PPO_Model_CartPole')
```

```python
model.save(PPO_Path)
```

```python
del model
```

```python
model = PPO.load(PPO_Path, env=env)
```

```python
model
```

### 5. Evaluation

##### Core metrics to look at:
- Average Reward
- Avg episode length

```python
env = gym.make('CartPole-v1', render_mode="human")
evaluate_policy(model, env, n_eval_episodes=10, render=True)
# env.close()
```

```python
env.close()
```

### 6. Test Model

```python
env = gym.make('CartPole-v1', render_mode="human")
episodes = 5
for episode in range(episodes):
    obs, _ = env.reset()
    done = False
    truncate = False
    score = 0

    while not done and not truncate:
        # env.render()
        action, _ = model.predict(obs)
        obs, reward, done, truncate, info = env.step(action)
        score += reward
    print(f"Episode:{episode+1} Score:{score}")
env.close()
```

### 8. Adding a callback to the training stage

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
save_path = "Saved Models"
```

```python
eval_env = gym.make('CartPole-v1')
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(
    eval_env,
    callback_on_new_best = stop_callback,
    eval_freq=10000,
    best_model_save_path = save_path,
    verbose=1
)
```

```python
model = PPO('MlpPolicy', eval_env, verbose=1, tensorboard_log = log_path)
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```

### 9. Changing Policies

```python
net_arch = [dict(pi=[128,128,128,128], vf=[128,128,128,128])]
```

```python
model = PPO('MlpPolicy', eval_env, verbose=1, tensorboard_log = log_path, policy_kwargs={'net_arch':net_arch})
```

```python
model.learn(total_timesteps=20000, callback=eval_callback)
```
